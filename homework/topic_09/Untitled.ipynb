```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np
```

```python
temperature = pd.read_csv('./temperature.csv', delimiter = ',')
Detroit = temperature['Detroit'][1:].values
```

```python
plt.hist(Detroit, bins = 100, density = True)
plt.show()
```
